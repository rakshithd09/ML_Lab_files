```python
from sklearn.preprocessing import LabelEncoder

id_train = df_train1.index.values
labels = df_train1['country_txt']

df_train1 = df_train1.drop('country_txt', axis=1, inplace=False)

df_train1 = pd.concat([labels,df_train1], axis=1, join='inner')

# Label encoding for the categorical data eg: ...Iraq -> 7, Libya -> 10...
le = LabelEncoder()
y = le.fit_transform(labels)
X = df_train1.drop('country_txt', axis=1, inplace=False)

X = X.drop('region_txt', axis=1, inplace=False)
X = X.drop('imonth', axis=1, inplace=False)
X = X.drop('iday', axis=1, inplace=False)
X = X.drop('iyear', axis=1, inplace=False)
```

```python
from sklearn import cross_validation
## Spliting of training dataset into 70% training data and 30% testing data randomly
features_train, features_test, labels_train, labels_test = cross_validation.train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
## Decision Tree 
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(features_train, labels_train)
prediction = clf.predict(features_test)
## Computing accuracy
from sklearn.metrics import accuracy_score
print (accuracy_score(prediction, labels_test))
```

```python
## Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(features_train, labels_train)
prediction = clf.predict(features_test)
## Computing accuracy
print (accuracy_score(prediction, labels_test))
```

```python
## SVM 
from sklearn import svm
clf = svm.SVC(kernel="rbf") 
clf.fit(features_train, labels_train)
prediction = clf.predict(features_test)
## Computing accuracy
print (accuracy_score(prediction, labels_test))
```
